```python
def na_perc(dataframe):
    output = round(dataframe.isnull().sum()/len(dataframe.index)*100,2)
    return output
```

```python
# Finding Missing values in columns

Na_columns = na_perc(input_1)
Na_columns
```

```python
# Finding columns with null values

Na_columns = Na_columns[Na_columns>0]
Na_columns
```

```python
# Correlation
```

```python
# Top 10 corelated variables: Target0 dataaframe

corr = Target0.corr()
corrdf = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))
corrdf = corrdf.unstack().reset_index()
corrdf.columns = ['Var1', 'Var2', 'Correlation']
corrdf.dropna(subset = ['Correlation'], inplace = True)
corrdf['Correlation'] = round(corrdf['Correlation'], 2)
corrdf['Correlation'] = abs(corrdf['Correlation'])
corrdf.sort_values(by = 'Correlation', ascending = False).head()
```

```python
# Top 10 correlated variables: Target1 dataframe

corr = Target1.corr()
corrdf = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))
corrdf = corrdf.unstack().reset_index()
corrdf.columns = ['Var1', 'Var2', 'Correlation']
corrdf.dropna(subset = ['Correlation'], inplace = True)
corrdf['Correlation'] = round(corrdf['Correlation'], 2)
corrdf['Correlation'] = abs(corrdf['Correlation'])
corrdf.sort_values(by = 'Correlation', ascending = False).head()
```

```python
#Observation:

# From the above corelation analysis, 
# we can say that the highest correlation (1.0) is between ENTRANCES_MEDI with ENTRANCES_AVG
```
